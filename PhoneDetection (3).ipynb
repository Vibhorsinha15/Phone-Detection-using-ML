```python
import os

val_labels = '/content/MergedDataset/labels/valid'
empty_or_missing = 0

for img in os.listdir('/content/MergedDataset/images/valid'):
    label_file = os.path.join(val_labels, img.replace('.jpg', '.txt').replace('.png', '.txt'))
    if not os.path.exists(label_file):
        print("Missing label:", img)
        empty_or_missing += 1
    else:
        with open(label_file) as f:
            content = f.read().strip()
            if not content:
                print("Empty label:", img)
                empty_or_missing += 1

print(f"Total invalid or missing label files: {empty_or_missing}")
```

```python
import os

# Path to validation label files
label_dir = '/content/MergedDataset/labels/valid'  # Change if your valid path differs

# Loop through all .txt files in the label directory
for fname in os.listdir(label_dir):
    if fname.endswith('.txt'):
        file_path = os.path.join(label_dir, fname)

        with open(file_path, 'r') as f:
            lines = f.readlines()

        # Keep only lines that start with class '0'
        filtered_lines = [line for line in lines if line.strip().startswith('0')]

        # Overwrite file with filtered content
        with open(file_path, 'w') as f:
            f.writelines(filtered_lines)

print("✅ Filtered validation labels to only keep class 0.")
```

```python
import pandas as pd
import os
import shutil

# Load labels.csv (assuming it's already extracted in /content)
df = pd.read_csv('/content/PhoneData/labels.csv')

# Create necessary directories
os.makedirs('/content/PhoneData/images/train', exist_ok=True)
os.makedirs('/content/PhoneData/labels/train', exist_ok=True)

# Convert to YOLO format and organize data
for _, row in df.iterrows():
    filename = row['filename']
    width = row['width']
    height = row['height']
    xmin = row['xmin']
    ymin = row['ymin']
    xmax = row['xmax']
    ymax = row['ymax']

    # Normalize bbox
    x_center = ((xmin + xmax) / 2) / width
    y_center = ((ymin + ymax) / 2) / height
    bbox_width = (xmax - xmin) / width
    bbox_height = (ymax - ymin) / height

    # Save YOLO label file
    label_path = f'/content/PhoneData/labels/train/{filename.replace(".jpg", ".txt")}'
    with open(label_path, 'w') as f:
        f.write(f"0 {x_center} {y_center} {bbox_width} {bbox_height}")

    # Copy image to /images/train
    src_img = f"/content/PhoneData/positive/{filename}"
    dst_img = f"/content/PhoneData/images/train/{filename}"
    if os.path.exists(src_img):
        shutil.copyfile(src_img, dst_img)
```

```python
import os, shutil
from sklearn.model_selection import train_test_split
from glob import glob

img_dir = '/content/PhoneData/images/train'
label_dir = '/content/PhoneData/labels/train'

# Get all image files
images = glob(f"{img_dir}/*.jpg") + glob(f"{img_dir}/*.png")

# Filter: keep only images with corresponding label files
valid_images = []
for img_path in images:
    fname = os.path.basename(img_path)
    label_path = os.path.join(label_dir, fname.replace('.jpg', '.txt').replace('.png', '.txt'))
    if os.path.exists(label_path):
        valid_images.append(img_path)

# Split into train and val
train_imgs, val_imgs = train_test_split(valid_images, test_size=0.2, random_state=42)

# Create val folders
os.makedirs('/content/PhoneData/images/val', exist_ok=True)
os.makedirs('/content/PhoneData/labels/val', exist_ok=True)

# Move val images + labels
for img_path in val_imgs:
    fname = os.path.basename(img_path)
    label_path = os.path.join(label_dir, fname.replace('.jpg', '.txt').replace('.png', '.txt'))

    shutil.move(img_path, f"/content/PhoneData/images/val/{fname}")
    shutil.move(label_path, f"/content/PhoneData/labels/val/{os.path.basename(label_path)}")

print("✅ Validation set created successfully!")
```

```python
import os
import random
import shutil

# Set seed for reproducibility
random.seed(42)
```

```python
base_path = '/content/PhoneData'
image_dir = os.path.join(base_path, 'images')
label_dir = os.path.join(base_path, 'labels')

# Output folders
train_images = os.path.join(base_path, 'images/train')
val_images = os.path.join(base_path, 'images/val')
train_labels = os.path.join(base_path, 'labels/train')
val_labels = os.path.join(base_path, 'labels/val')

# Create directories
for folder in [train_images, val_images, train_labels, val_labels]:
    os.makedirs(folder, exist_ok=True)
```

```python
images = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png', '.jpeg'))]
random.shuffle(images)

split_idx = int(0.8 * len(images))
train_files = images[:split_idx]
val_files = images[split_idx:]

def move_files(file_list, src_img, src_lbl, dest_img, dest_lbl):
    for file in file_list:
        name, _ = os.path.splitext(file)
        img_src = os.path.join(src_img, file)
        lbl_src = os.path.join(src_lbl, f'{name}.txt')

        img_dst = os.path.join(dest_img, file)
        lbl_dst = os.path.join(dest_lbl, f'{name}.txt')

        if os.path.exists(img_src) and os.path.exists(lbl_src):
            shutil.copy(img_src, img_dst)
            shutil.copy(lbl_src, lbl_dst)

move_files(train_files, image_dir, label_dir, train_images, train_labels)
move_files(val_files, image_dir, label_dir, val_images, val_labels)
```

```python
import pandas as pd
import os

# Load CSV
csv_path = '/content/PhoneData/labels.csv'
df = pd.read_csv(csv_path)

# Ensure labels directory exists
labels_dir = '/content/PhoneData/labels/positive'
os.makedirs(labels_dir, exist_ok=True)

# Mapping class "Comp" to 0
class_map = {'Comp': 0}

# Create YOLO label files
for _, row in df.iterrows():
    filename = row['filename']
    image_width = row['width']
    image_height = row['height']

    x_center = ((row['xmin'] + row['xmax']) / 2) / image_width
    y_center = ((row['ymin'] + row['ymax']) / 2) / image_height
    width = (row['xmax'] - row['xmin']) / image_width
    height = (row['ymax'] - row['ymin']) / image_height
    class_id = class_map[row['class']]

    label_filename = os.path.splitext(filename)[0] + '.txt'
    label_path = os.path.join(labels_dir, label_filename)

    with open(label_path, 'a') as f:
        f.write(f"{class_id} {x_center:.6f} {y_center:.6f} {width:.6f} {height:.6f}\n")
```

```python
import shutil

base_path = '/content/PhoneData'
os.makedirs(f'{base_path}/images/train', exist_ok=True)
os.makedirs(f'{base_path}/images/val', exist_ok=True)
os.makedirs(f'{base_path}/labels/train', exist_ok=True)
os.makedirs(f'{base_path}/labels/val', exist_ok=True)

# Move all positive and negative images to train/val (80-20 split)
from sklearn.model_selection import train_test_split

pos_images = [f for f in os.listdir(f'{base_path}/positive') if f.endswith('.jpg')]
neg_images = [f for f in os.listdir(f'{base_path}/negative') if f.endswith('.jpg')]

pos_train, pos_val = train_test_split(pos_images, test_size=0.2, random_state=42)
neg_train, neg_val = train_test_split(neg_images, test_size=0.2, random_state=42)

def move_files(image_list, src_img, src_lbl, dst_img, dst_lbl):
    for img in image_list:
        shutil.copy(os.path.join(src_img, img), os.path.join(dst_img, img))
        label_file = os.path.splitext(img)[0] + '.txt'
        if os.path.exists(os.path.join(src_lbl, label_file)):
            shutil.copy(os.path.join(src_lbl, label_file), os.path.join(dst_lbl, label_file))

move_files(pos_train, f'{base_path}/positive', f'{base_path}/labels/positive', f'{base_path}/images/train', f'{base_path}/labels/train')
move_files(pos_val, f'{base_path}/positive', f'{base_path}/labels/positive', f'{base_path}/images/val', f'{base_path}/labels/val')
move_files(neg_train, f'{base_path}/negative', f'{base_path}/labels/negative', f'{base_path}/images/train', f'{base_path}/labels/train')
move_files(neg_val, f'{base_path}/negative', f'{base_path}/labels/negative', f'{base_path}/images/val', f'{base_path}/labels/val')
```

```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load model2
model2 = YOLO('/content/runs/detect/phone-detect/weights/best.pt')  # Use the correct path for model2

# List of video files to process
video_files = ['video1.mp4', 'video2.mp4']  # Add both video files here

# Loop through each video file
for video_path in video_files:
    # Open the video
    cap = cv2.VideoCapture(video_path)

    # Get video properties
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

    # Prepare for saving the output video
    output_path = f'/content/output_{video_path.split("/")[-1]}'  # Name the output file based on the input video
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

    # Process each frame of the video
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Perform inference on the frame using model2
        results = model2.predict(frame)

        # Get the image with predictions
        img = results[0].plot()  # Image with bounding boxes drawn

        # Write the frame with bounding boxes to the output video
        out.write(img)

        # Display the frame with predictions (optional)
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()

    # Release resources for the current video
    cap.release()
    out.release()

    print(f"Video processing for {video_path} is complete. Output saved at: {output_path}")
```

```python
import cv2
from ultralytics import YOLO

# Load your trained model
model = YOLO('runs/detect/phone-detect2/weights/best.pt')  # Adjust path if needed

# Open the default camera (0 is usually the built-in webcam)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run prediction
    results = model.predict(frame)

    # Visualize predictions
    annotated_frame = results[0].plot()

    # Show output
    cv2.imshow("Phone Detection - Live", annotated_frame)

    # Exit on pressing 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release and destroy
cap.release()
cv2.destroyAllWindows()
```
